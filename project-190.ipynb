```python
from collections import defaultdict
def group_anagrams(a):
    dfdict = defaultdict(list)
    for i in a:
        sorted_i = " ".join(sorted(i))
        dfdict[sorted_i].append(i)
    return dfdict.values()

words =["eat", "ate", "bat", "tea", "arc", "car"]
group_anagrams(words)
```

```python
# Given an array cotaining the range of numbers from 0 to n with a missing number,
#find the missing number in the array

def findMissingNumbers(n):
    numbers = set(n)
    length = len(n)
    output = []
    for i in range(1, n[-1]):
        if i not in numbers:
            output.append(i)
    return output
    
listOfNumbers = [1, 2, 3, 5, 6, 7, 9, 10, 13]
print(findMissingNumbers(listOfNumbers))
```

```python
#Grouping elements of same indices
input = [[10, 20, 30], [40, 50, 60], [70, 80, 90]]
output = []
index = 0

for i in range(len(input[0])):
    output.append([])
    for j in range(len(input)):
        output[index].append(input[j][index])
    index = index + 1
    
a,b,c = output[0], output[1], output[2]
print(a, b, c)
```

```python
list = [12, 16, 20, 20, 12, 30, 25, 23, 24, 20]
frequency = {}
for i in list:
    frequency.setdefault(i, 0)
    frequency[i] += 1
    
frequent = max(frequency.values())
for i,j in frequency.items():
    if j == frequent:
        mode = i
        
print(mode)
```

```python
#Sets in python
#We can add tuples, but we cannot add dictionaries or lists in a set
first_set = {1, 2, 3}
print(first_set)

#Create set using set() constructor
empty_set = set()
print(empty_set)

first_set = set((1, 2, 3)) #converting the tuple into set
print(first_set)

#add elements in a set using add() method
first_set.add(4)
print(first_set)

#we can add more than one element in a set with update() method
first_set.update([5, 6, 7])
print(first_set)

first_set.update(['ABC'])
print(first_set)

first_set.update('ABC')
print(first_set)

#If we want to delete elements from sets, we can use methods discard() or remove()
first_set.remove('ABC')
print(first_set)

first_set.remove(4)
print(first_set)

#first_set.remove(100)
first_set.discard(100)

#We can find the union of two sets using the pipe operator | or the union() method
first_set = {1, 2}
second_set = {2, 3, 4}

union_set = first_set.union(second_set)
print(union_set)

#We can find intersection using operator & or the intersection() method:
intersection_set = first_set & second_set
print(intersection_set)

'''
We can find the difference between two sets using the - operator or the method difference()
The difference between two sets represents the collection that contains only the elements that are in the first set, but not in the second
'''
difference_set_first = first_set.difference(second_set)
print(difference_set_first)

difference_set_second = second_set - first_set
print(difference_set_second)

```
