```python
from collections import defaultdict
def group_anagrams(a):
    dfdict = defaultdict(list)
    for i in a:
        sorted_i = " ".join(sorted(i))
        dfdict[sorted_i].append(i)
    return dfdict.values()

words =["eat", "ate", "bat", "tea", "arc", "car"]
group_anagrams(words)
```

```python
# Given an array cotaining the range of numbers from 0 to n with a missing number,
#find the missing number in the array

def findMissingNumbers(n):
    numbers = set(n)
    length = len(n)
    output = []
    for i in range(1, n[-1]):
        if i not in numbers:
            output.append(i)
    return output
    
listOfNumbers = [1, 2, 3, 5, 6, 7, 9, 10, 13]
print(findMissingNumbers(listOfNumbers))
```

```python
#Sets in python
#We can add tuples, but we cannot add dictionaries or lists in a set
first_set = {1, 2, 3}
print(first_set)

#Create set using set() constructor
empty_set = set()
print(empty_set)

first_set = set((1, 2, 3)) #converting the tuple into set
print(first_set)

#add elements in a set using add() method
first_set.add(4)
print(first_set)

#we can add more than one element in a set with update() method
first_set.update([5, 6, 7])
print(first_set)

first_set.update(['ABC'])
print(first_set)

first_set.update('ABC')
print(first_set)

#If we want to delete elements from sets, we can use methods discard() or remove()
first_set.remove('ABC')
print(first_set)

first_set.remove(4)
print(first_set)

#first_set.remove(100)
first_set.discard(100)

#We can find the union of two sets using the pipe operator | or the union() method
first_set = {1, 2}
second_set = {2, 3, 4}

union_set = first_set.union(second_set)
print(union_set)

#We can find intersection using operator & or the intersection() method:
intersection_set = first_set & second_set
print(intersection_set)

'''
We can find the difference between two sets using the - operator or the method difference()
The difference between two sets represents the collection that contains only the elements that are in the first set, but not in the second
'''
difference_set_first = first_set.difference(second_set)
print(difference_set_first)

difference_set_second = second_set - first_set
print(difference_set_second)

```

```python
class Article:
    blog = 'https://www.python.org/'

    # the init method is called when an instance of the class is created
    def __init__(self, title, content):
        self.title = title
        self.content = content

    @classmethod
    def get_blog(cls):
        return cls.blog

    @staticmethod
    def print_creation_date(date):
        print(f'The blog was created on {date}')


article = Article('First Article', 'This is the first article')

# Calling the static method using the object
article.print_creation_date('2022-07-18')  # The blog was created on 2022-07-18

# Calling the static method using the class name
Article.print_creation_date('2022-07-21')  # The blog was created on 2022-07-21
```

```python
'''
Private attributes can be accessed directly only from inside the class.
We can make properties attributes by using the double underscore, as you can see in
the following example.
we are declaring private methods using the double underscore in front of the name 
of the method that we want to make private: The following code snippet is for code error
'''
class Motorbike:
    def __init__(self, company, color, old):
        self.company = company
        self.color = color
        self.__old = old  # This is private property
        
    def __get_description(self):
        return self.__old


# The following will throw an error because private method and private attributes are 
# called outside class
bike = Motorbike('Hero', 'Black', True)
print(bike)
print(bike.__get_description)
```

```python
'''
We are then supposed to write public methods that can provide access to the outside world.
These methods are called getters and setters.
A getter method is a method that we use to get the value of an attribute.
A setter is a method that we use to set the value of an attribute.
Let's define first define a getter and a setter method that we can use to get values:
'''

class Smartphone:
    def __init__(self, type=None): # type is for no argument
        self.__type = type # Setting the type in the beginning when object is created
        
    def set_type(self, value):
        self.__type = value
        
    def get_type(self):
        return self.__type

    
smartphone = Smartphone("Iphone") # Using the constructoo method
print(smartphone.get_type())
smartphone.set_type("Samsung")
print(smartphone.get_type())
```

```python
'''
What we have done so far is set and also read the value of a private attribute of an object created from 
the Smartphone class. We can also define getters and setters using the @property annotation.
'''


class Bicycle:
    def __init__(self, manufacture, color, is_old):
        self._manufacture = manufacture
        self._color = color
        self._is_old = is_old

    @property
    def manufacture(self):
        return self._manufacture

    @manufacture.setter
    def manufacture(self, manufacture):
        self._manufacture = manufacture

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, color):
        self._color = color

    @property
    def is_old(self):
        return self._is_old

    @is_old.setter
    def is_old(self, is_old):
        self._is_old = is_old

    # Classes can also have other methods as well, and not just getters and setters
    def get_description(self):
        desc = f"This color is{self._color} and the company is {self._manufacture} where bike is too ole {self._is_old}"
        return desc


bike = Bicycle("Bajaj", "Grey", True)
print(bike.manufacture)
print(bike.color)
print(bike.is_old)
print(bike.get_description())

'''
We can also modify the value that we initially used for any attribute by simply typing the name of the object
and the attribute that we want to modify:
'''
bike.color = 'black'
bike.is_old = False


print(bike.color)
print(bike.is_old)
print(bike.get_description())
```

```python
# Inheritance
class Book:
    def __init__(self, name, author):
        self.name = name
        self.author = author
        
    def short_paragraph(self):
        paragraph = "This is a short paragraph"
        return paragraph
    
    
book1 = Book("Sapiens", "Harari")
print(book1.name)
print(book1.author)
print(book1.short_paragraph())

# creating the subclass of the parent class Book
class BookDetails(Book):
    def __init__(self, name, author):
        Book.__init__(self, name, author)  # Here we call the constructor of the parent class Book
    
    def get_description(self):
        description = "Name: " + self.name + ". "
        description += "Author: " + self.author
        return description
 
book2 = BookDetails("behave", "Robert Sapolsky")
print(book2.get_description())
print(book2.short_paragraph())
```

```python
'''
Types of Inheritance
We can have different types of inheritance based on the relationship of parent classes and child classes:
    Single
    Multi-level
    Hierarchical
    Multiple
    Hybrid
'''
# Single: We can have a class that inherits only from another class
class Animal:
    def __init__(self):
        self.health = 100
    
    def get_health(self):
        return self.health
    
class Cat(Animal):
    def __init__(self, name):
        super().__init__()
        self.name = name
        self.health = 200
        
    def move(self):
        print("See!, The cat can move ")
        
cat = Cat("cat")

# calling the method from parent class
print(cat.get_health())
#calling the method from child class
cat.move() # If we print(cat.move()), it will give the None output too 


'''
Multi-level inheritance: 
    This is another type of inheritance where a class inherits from another class which 
    inherits from another class: Class A inherits from Class B which inherits from Class C
'''
class Creature:
    def __init__(self, alive):
        self.alive = alive
        
    def is_alive(self):
        return self.alive
    
class Animal(Creature):
    def __init__(self):
        super().__init__(True)
        self.health = 300
            
    def get_health(self):
        return self.health
    
class Cat(Animal):
    def __init__(self, name):
        super().__init__()
        self.name = name
        
    def move(self):
        print("Look the cat can move easily: ", self.name)
        
cat = Cat("LoudCat")

# Calling the method from the parent of the parent class
print(cat.is_alive())

# Calling the method from the parent class
print(cat.get_health())

#calling the method from the child class
cat.move()
            
'''
Hierarchical inheritance:
    When we derive multiple child classes from the same parent class, then we have 
    hierarchical inheritance. These child classes inherit from the parent class.
'''
class Location:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def get_location(self):
        return self.x, self.y
    
class Continent(Location):
    pass

class Country(Location):
    pass

continent = Continent(10, 20)
country = Country(100, 300)

print(continent.get_location())
print(country.get_location())


'''
Multiple inheritance:
    We can have another type of inheritance, namely multiple inheritance which can help us 
    inherit from more than one class at the same time.Let's assume that we have a class called 
    Date and another one called Time. We can then implement another class then inherits from
    both classes
'''

class Date:
    date = "2022-07-22" #hardcoded date
    
    def get_date(self):
        return self.date
    

class Time:
    time = "06:30:45" #hardcoded time
    
    def get_time(self):
        return self.time
    
# Inheriting from the both the class  
class DateTime(Date, Time):
    def get_date_time(self):
        return f"The date is: {self.get_date()} and the time is {self.get_time()}"
    
date_time = DateTime()
date_time.get_date_time()


'''
Hybrid inheritance:
    Hybrid inheritance is a combination of multiple and multi-level inheritance:
'''
class Vehicle:
    def print_vehicle(self):
        print("vehicle")
        
class Car(Vehicle):
    def print_car(self):
        print("car")
        
class Ferrari(Car):
    def print_ferrari(self):
        print("ferrari")
        
class Driver(Ferrari, Car):
    def print_driver(self):
        print("driver")
        
        
driver = Driver()

# Calling all the method from the subclass
print(driver.print_vehicle())
print(driver.print_car())
print(driver.print_ferrari())
driver.print_driver() # If we print this statement it will also give a none value

    
    
    
```

```python
'''
Polymorphism: 
    We can have polymorphism using both methods and inheritance. Refers to the possibility
    of an object behaving like different forms and calling different behaviors. An example of
    a built-in function that uses polymorphism is the method len() which can be used for
    both strings and lists
'''
# Polymorphism using methods
class Condo:
    def __init__(self, area):
        self.area = area
        
    def gat_area(self):
        return self.area
    
class Apartment:
    def __init__(self, area):
        self.area = area
        
    def get_area(self):
        return self.area
    
# create objects using two classes
condo = Condo(100)
apartment = Apartment(200)

# Now, we can put both of them in a list and call the same method for both objects
places_to_live = [condo, apartment]
for places in places_to_live:
    print(places.get_area())  # same method for bothe objects
    

```

```python
def find_duplicates(x):
    length = len(x)
    duplicate = []
    for i in range(length):
        n = n+1
        for a in range(n, length):
            if x[i] == x[a] and x[i] not in duplicate
            duplicate.append(x[i])
    return duplicate

names = ["paskal", "Aman", "Akanksha", "paskal", "Aman", "Dikshya"]
print(find_duplicates(names))
```

```python
# Netflix Recommendation System

import os
import pandas as pd
import numpy as np
from sklearn.feature_extraction import text
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv("/kaggle/input/netflix-movies-and-tv-shows-2021/netflixData.csv")
print(data.head())

# let’s have a look at whether the data contains null values or not
print(data.isnull().sum())

'''
The dataset contains null values, but before removing the null values,
let’s select the columns that we can use to build a Netflix recommendation system
'''

data = data[["Title", "Description", "Content Type", "Genres"]]
print(data.head())

data = data.dropna()
print(data.isnull().sum())


# Now I will clean the Title column as it contains some data preparation

import nltk
import re
nltk.download("stopwords")
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword = set(stopwords.words("english"))


def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text = " ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text
data['Title'] = data["Title"].apply(clean)


print(data.Title.sample(10))

# Now I will use the Genres column as the feature to recommend similar
# content to the user

feature = data["Genres"].tolist()
tfidf = text.TfidfVectorizer(input=feature, stop_words='english')
tfidf_matrix = tfidf.fit_transform(feature)
similarity = cosine_similarity(tfidf_matrix)

'''Now I will set the Title column as an index so that we can find 
similar content by giving the title of the movie or TV show as an input
'''
indices = pd.Series(data.index,
                   index=data['Title']).drop_duplicates()

# Now here’s how to write a function to recommend Movies and TV shows on Netflix:
def netFlix_recommendation(title, similarity = similarity):
    index = indices[title]
    similarity_scores = list(enumerate(similarity[index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores = similarity_scores[0:10]
    movieindices = [i[0] for i in similarity_scores]
    return data['Title'].iloc[movieindices]


print(netFlix_recommendation("girlfriend"))
```
