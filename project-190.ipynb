```python
from collections import defaultdict
def group_anagrams(a):
    dfdict = defaultdict(list)
    for i in a:
        sorted_i = " ".join(sorted(i))
        dfdict[sorted_i].append(i)
    return dfdict.values()

words =["eat", "ate", "bat", "tea", "arc", "car"]
group_anagrams(words)
```

```python
# Given an array cotaining the range of numbers from 0 to n with a missing number,
#find the missing number in the array

def findMissingNumbers(n):
    numbers = set(n)
    length = len(n)
    output = []
    for i in range(1, n[-1]):
        if i not in numbers:
            output.append(i)
    return output
    
listOfNumbers = [1, 2, 3, 5, 6, 7, 9, 10, 13]
print(findMissingNumbers(listOfNumbers))
```

```python
#Grouping elements of same indices
input = [[10, 20, 30], [40, 50, 60], [70, 80, 90]]
output = []
index = 0

for i in range(len(input[0])):
    output.append([])
    for j in range(len(input)):
        output[index].append(input[j][index])
    index = index + 1
    
a,b,c = output[0], output[1], output[2]
print(a, b, c)
```

```python
list = [10, 20, 30, 20, 10, 60, 80, 20, 30, 10, 20, 80, 90, 100]
list.sort()
if len(list) % 2 == 0:
    m1 = list[len(list)//2]
    m2 = list[len(list)//2 -1]
    median = (m1 + m2)/2
else:
    median = list[len(list)//2]
print(median)
```

```python
list = [12, 16, 20, 20, 12, 30, 25, 23, 24, 20]
frequency = {}
for i in list:
    frequency.setdefault(i, 0)
    frequency[i] += 1
    
frequent = max(frequency.values())
for i,j in frequency.items():
    if j == frequent:
        mode = i
        
print(mode)
```

```python
#SequenceMatcher in Python
from difflib import SequenceMatcher
text1 = "My name is Paskal Sunari"
text2 = "Hi, my name is paskal sunari"
sequenceScore = SequenceMatcher(None, text1, text2).ratio()
print(f"Both are {sequenceScore*100}% similar")
```

```python
#list comprehension
numbers = [1 , 0, -1, -3, -4, 4, 5]
positive_number = [number * 10 for number in numbers]
print(positive_number)
```

```python
numbers = [1 , 0, -1, -3, -4, 4, 5]
new_number = []

for number in numbers:
    if number > 0:
        new_number.append(number + 100)

print(new_number)
```

```python
#List comprehension fo above program
numbers = [1 , 0, -1, -3, -4, 4, 5]

positive = [number + 100 for number in numbers if number >0]

print(positive)
```

```python
#list comprehension for multiple list

first_list = [1, 2, 3]
second_list = [50]

double_list = [first_element + second_element
               for first_element in first_list for second_element in second_list]

print(double_list)
```
